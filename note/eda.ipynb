```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
edf = pd.read_parquet('../data/extract/')
```

```python
edf.head(3)
```

```python
edf_grouped = edf.groupby('openDt').size().reset_index(name='count')
```

```python
plt.plot(edf_grouped['openDt'], edf_grouped['count'])
plt.xlabel('open_dt')
plt.ylabel('cnt')
plt.title('open_dt cnt')
plt.xticks(edf_grouped['openDt'][::15], rotation=90)
plt.show()
```
